This example shows how to perform a auto-correlation using scipy library.

```python
#Import necessary libraries

import numpy as np

from scipy import signal

import matplotlib.pyplot as plt
```

1. Generate signals we want to analyse.

We choose a step function and a decaying exponent function of 300 samples each.

```python
sig = np.repeat([0., 1., 0.], 100)

win = np.zeros(300)
win[100:] = np.exp(np.linspace(5,0,200))

plt.plot(sig)
plt.show()
plt.plot(win, color='red')
```

2. We porform auto-correlation of the step function using function correlate() from the scipy library.

```python
filtered = signal.correlate(sig, sig, mode='same') / sum(sig)

plt.plot(filtered)
```

3. We porform auto-correlation of the exponent function.

```python
filtered = signal.correlate(win, win, mode='same') / sum(win)

plt.plot(filtered, color='r')
```
